```python
# Get input from the user for capacity and department
N = int(input("Ingrese el cupo maximo por departamento: "))
departamento_seleccionado = input("Ingrese el departamento: ")

# definición de una funcion  para estimar puntaje de admisión
def calcular_puntaje_admision(linea, opcion):
  puntaje = 0
  departamento = linea[opcion]
  if departamento == "Physics":
    puntaje = (linea["EPH"] + linea["EMA"] +linea["ESP"]) /3
  elif departamento == "Mathematics":
    puntaje = (linea["ECH"] + linea["ECS"]) + linea["ESP"] /3
  elif departamento == "Engineering":
    puntaje = (linea["EMA"] + linea["ECS"] + linea["EPH"] + linea["ESP"]) /4
  elif departamento == "Chemistry":
    puntaje = (linea["ECH"] + linea["EPH"]) + linea["ESP"] /3
  elif departamento == "Biotech":
    puntaje = (linea["EMA"] + linea["ECS"] + linea["ECH"] +linea["ESP"]) /4
  else:
    return None
  return round(puntaje, 2)

# Filter applicants who selected the chosen department as one of their options
df_filtered = df[(df['OP1'] == departamento_seleccionado) | (df['OP2'] == departamento_seleccionado) | (df['OP3'] == departamento_seleccionado)].copy()

# Calculate the score for the selected department for each filtered applicant
def calcular_puntaje_departamento(row, departamento):
    if row['OP1'] == departamento:
        return calcular_puntaje_admision(row, 'OP1')
    elif row['OP2'] == departamento:
        return calcular_puntaje_admision(row, 'OP2')
    elif row['OP3'] == departamento:
        return calcular_puntaje_admision(row, 'OP3')
    return None


df_filtered['Puntaje_Departamento'] = df_filtered.apply(lambda row: calcular_puntaje_departamento(row, departamento_seleccionado), axis=1)

display(df_filtered)
```

```python
df_sorted = df_filtered.sort_values(by='Puntaje_Departamento', ascending=False)
display(df_sorted)
```
